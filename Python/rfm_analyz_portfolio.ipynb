```python
#проверяю, что за мероприятия, у которых не указано event name. Оказалось, что это обычные мероприятия, у которых по каким-то причинам в базе указан только event type без event name, поэтому оставляю эти данные
data_not_eventname = data[data['event name'].isna()]
data_not_eventname
```

```python
#удаляю билеты, оформленные администратором (они для блогеров, артистов и т.п.)
data = data[data['email']!='moil@vmestesmomoy.re']
```

```python
#удаляю билеты c суммой 0, оформленные как "подвешенные", т.е. бесплатные для тех гостей, кто в сложном положении и попросил такой билет 
data = data[data['sum']!=0]
```

```python
#ищу самую позднюю покупку для того, чтобы от нее вести отсчет для показателя R
data['date'].max()
```

```python
import datetime as dt

#создаю переменную, в которую кладу максимальную дату
now = dt.datetime(2019,6,30)
```

```python
#создаю таблицу и рассчитываю 3 показателя - R(как давно покупал), F(количество покупок), M(сумму покупок)

rfm_data = data.groupby('email').agg({'date': lambda x: (now - x.max()).days,
                                        'ticket count': lambda x: len(x),
                                        'sum': lambda x: x.sum()})

#перевожу число дней в числовой формат
rfm_data['date'] = rfm_data['date'].astype(int)

#переименую столбцы
rfm_data.rename(columns={'date': 'recency', 
                         'ticket count': 'frequency', 
                         'sum': 'monetary_value'}, inplace=True)

rfm_data
```

```python
#нахожу значения квантилей для дальнейшего распеределения по группам
quantiles = rfm_data.quantile(q=[0.25,0.5,0.75])
quantiles
```

```python
#создаю словарь со значениями квантилей
quantiles = quantiles.to_dict()
quantiles
```

```python
#создаю новую таблицу для сегментации клиентов
rfm_segmentation = rfm_data

#функция для делений на группы по показателю R - чем меньше число (количество дней с последней покупки), тем лучше клиент
def rclass(x,p,d):
    if x <= d[p][0.25]:
        return 1
    if x <= d[p][0.50]:
        return 2
    if x <= d[p][0.75]:
        return 3
    else:
        return 4
    
#функция для делений на группы по показателям F и M - чем больше число (количество покупок и сумма покупок), тем лучше покупатель
def fmclass(x,p,d):
    if x <= d[p][0.25]:
        return 4
    if x <= d[p][0.50]:
        return 3
    if x <= d[p][0.75]:
        return 2
    else:
        return 1
```

```python
#добавляю новые столбцы в таблицу и применяю для них функции
rfm_segmentation['R_Quartile'] = rfm_segmentation['recency'].apply(rclass, args=('recency', quantiles,))
rfm_segmentation['F_Quartile'] = rfm_segmentation['frequency'].apply(fmclass, args=('frequency', quantiles,))
rfm_segmentation['M_Quartile'] = rfm_segmentation['monetary_value'].apply(fmclass, args=('monetary_value', quantiles,))
rfm_segmentation
```

```python
#создаю столбец, в котором объединяю показатели из 3х столбцов в 1
rfm_segmentation['RFM_class'] = rfm_segmentation['R_Quartile'].map(str) + rfm_segmentation['F_Quartile'].map(str) + rfm_segmentation['M_Quartile'].map(str)

rfm_segmentation
```

```python
#создаю отдельный df с итогами, сортирую по возрастанию, чтобы дальше его визуализировать

df_final_rfm = rfm_segmentation['RFM_class'].value_counts().reset_index()
df_final_rfm.columns = ['RFM_class', 'Count']

df_sorted = df_final_rfm.sort_values('Count', ascending=True)
df_sorted
```

```python
import matplotlib.pyplot as plt
# Создаю график, чтобы показать деление клиентов по сегментам
plt.figure(figsize=(12, 8))
bars = plt.barh(df_sorted['RFM_class'], df_sorted['Count'], 
                color='lightsalmon', edgecolor='maroon', alpha=0.7)

for bar in bars:
    width = bar.get_width()
    plt.text(width + 5, bar.get_y() + bar.get_height()/2, 
             f'{int(width)}', ha='left', va='center', fontsize=9)

plt.xlabel('Количество клиентов')
plt.ylabel('RFM-сегмент')
plt.title('Количество клиентов по RFM-сегментам\n(1 - лучшая оценка)', fontsize=14, fontweight='bold')
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import numpy as np

# Создаю тепловые карты для каждой группы сегментов
rfm_matrix = np.zeros((4, 4, 4))  # R, F, M

for _, row in df_final_rfm.iterrows():
    rfm = row['RFM_class']
    r, f, m = int(rfm[0])-1, int(rfm[1])-1, int(rfm[2])-1
    rfm_matrix[r, f, m] = row['Count']

fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Тепловая карта распределения клиентов по RFM-сегментам\n(1 - лучшая оценка)', 
             fontsize=16, fontweight='bold')

for r in range(4):
    ax = axes[r//2, r%2]
    sns.heatmap(rfm_matrix[r, :, :], 
                annot=True, fmt='.0f', 
                cmap='Oranges',
                cbar_kws={'label': 'Количество клиентов'},
                ax=ax)
    ax.set_title(f'Recency = {r+1} ({"Недавно" if r+1==1 else "Давно" if r+1==4 else "Средне"})')
    ax.set_xlabel('Monetary (Сумма)')
    ax.set_ylabel('Frequency (Частота)')
    ax.set_xticklabels(['1 (Много)', '2', '3', '4 (Мало)'])
    ax.set_yticklabels(['1 (Часто)', '2', '3', '4 (Редко)'])

plt.tight_layout()
plt.show()
```

```python
# Создаю функцию для определения качества сегмента
def get_segment_quality(rfm):
    r, f, m = int(rfm[0]), int(rfm[1]), int(rfm[2])
    score = r + f + m  # Сумма трех показателей (1-4 каждый)
    
    if score <= 4:
        return 'Лучшие (1-4)'
    elif score <= 7:
        return 'Хорошие (5-7)'
    elif score <= 10:
        return 'Средние (8-10)'
    else:
        return 'Слабые (11-12)'

# 2. Применяю функцию к каждому RFM-сегменту
df_final_rfm['Segment_Quality'] = df_final_rfm['RFM_class'].apply(get_segment_quality)
df_final_rfm
```

```python
#Группирую по сегментам качества
quality_summary = df_final_rfm.groupby('Segment_Quality')['Count'].sum()
quality_summary
```

```python
# Строю круговую диаграмму для визуализации сегментов по качеству
plt.figure(figsize=(6, 4))
wedges, texts, autotexts = plt.pie(quality_summary.values, 
                                   labels=quality_summary.index,
                                   autopct='%1.1f%%',
                                   colors=['lightsalmon', 'darksalmon', 'salmon', 'peachpuff'],
                                   startangle=90,
                                   textprops={'fontsize': 12})

for autotext in autotexts:
    autotext.set_color('black')
    autotext.set_fontweight('bold')

plt.title('Распределение клиентов по качеству RFM-сегментов', 
          fontsize=12, fontweight='bold')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

**Выводы**

RFM-сегментация составлена за период с 1 января 2019 г. по 30 июня 2019г, то есть за половину музыкального сезона.

Несколько специфичных моментов о досуговом бизнесе для родителей с маленькими детьми (от 1 мес до 3-4 лет):
- 1 билет покупается на семью, внутри билета осуществляется наполнение по количеству зрителей (мама, папа, дети). Сумма покупки зависит от полноты семьи и не сильно поддаётся внешнему маркетинговому воздействию.
- "выход в свет" на концерты и спектакли семьи с маленькими детьми осуществляют 1-2 раза за сезон, поэтому показатель F (частота) в данном анализе не совсем информативен.  
Поэтому при проведение RFM-анализа в классических розничных продажах для каждой группы (111, 112, 113 и т.д. ) был бы составлен отдельный план по удержанию, активации и т.д. А в данном случае в основном важен показатель R (давность покупки), исходя из которого и сделаны выводы.

Основные выводы:
1. В целом клиентская база выглядит здоровой и распределена, как 30% / 50% / 30% - самые активные / средние / слабые. 
2. Клиенты, которые приходили в течение 2х последних месяцев, составляют в сумме 374 семьи (26.8% ), из них 200 семей и 14.3% - лучшие; 174 семьи и 12,5% - хорошие. Им нужно анонсировать осенние мероприятия и начинать активно приглашать с сентября.
2. 665 семей (47,6%), которые приходили в течение последних 3х месяцев, необходимо активно звать на проводящиеся в данный момент летние мероприятия и анонсировать начало нового музыкального сезона с сентября.
3. 358 семей (25,6%), которые приходили только в начале года (более 4,5 месяцев назад), необходимо реактивировать и предложить им спец.предложение, которое действует, например, на все летние мероприятия.
